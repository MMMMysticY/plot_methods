# matplotlib 入门
matplotlib的图是画在**figure**上, 每个figure包含了一个或多个**axes**(一个可以指定坐标系的子区域)


```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

## 一个最简单的绘图例子
plt.subplots()和plt.plot() 默认直接绘制折线图

```python
fig, ax = plt.subplots() # 创建一个包含axes的figure
ax.plot([1,2,3,4],[1,4,2,3]); # plot直接绘制折线图 分别由x的list和y的list
```

```python
plt.plot([1,2,3,4], [1,4,2,3]); # 如果没有显式指定ax就直接plot 会自动创建一个ax进行折线图绘制
```

## 绘图的两种接口
1. 显式创建figure和axes 成为object-oriented style面向对象风格
2. 依赖pyplot自动创建figure和axes

```python
x = np.linspace(0,2,100)

## 面向对象的方法 显式定义了figure和ax
fig, ax = plt.subplots()
ax.plot(x, x, label = 'linear')
ax.plot(x, x**2, label = 'quadratic')
ax.plot(x, x**3, label = 'cubic')
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')
ax.set_title('Simple Plot')
ax.legend()
plt.show()
```

```python
# 不断采用plt的函数 面向过程进行绘图
plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

## 通用绘图模板

```python
# step1 准备数据
x = np.linspace(0, 2, 100)
y = x ** 2

# step2 设置绘图样式 (非必须在这里设置 可以再绘图细节部分进行设置)
mpl.rc('lines', linewidth = 4, linestyle = '-.')

# step3 定义布局
fig, ax = plt.subplots()

# step4 绘制图像
ax.plot(x, y, label = 'line')

# step5 添加标签 文字 图例
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend() ;
```
