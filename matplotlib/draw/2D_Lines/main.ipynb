```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_dir/raw_size.csv', header=None)
data.columns = ['raw_size', 'now_size','1','2','3','4']
data = data[['raw_size', 'now_size']]
data
```

```python
data.describe()
```

```python
data.sort_values(by = 'raw_size', ascending=True, inplace=True)
```

```python
data['raw_size'].values.shape
```

```python
data = data.sample(frac=1)
```

```python
data.to_csv('data_dir/raw_size_1.csv', header=None)
```

```python
x = range(0,1000)
y1 = data['raw_size'].values
y2 = data['now_size'].values

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
plt.show();
```
